```python
import json
import chess.pgn
import os
import re
import pandas as pd
import pickle
```

```python
def get_file_list_json(local_path):
    tree = os.walk(str(local_path))
    file_list = []
    out = []
    test = r'.+json$'
    for i in tree:
        file_list = i[2]

    for name in file_list:
        if(len(re.findall(test, name))):
            out.append(local_path +'\\'+ name)
    return out
```

```python
inp_dirs = [r'C:\Users\Giorgio\Documents\Python Scripts\Scacchi chessbase DB\data\chessbase\New Database_1000_1979',
           r'C:\Users\Giorgio\Documents\Python Scripts\Scacchi chessbase DB\data\chessbase\New Database_1980_1999',
           r'C:\Users\Giorgio\Documents\Python Scripts\Scacchi chessbase DB\data\chessbase\New Database_2000_2006',
           r'C:\Users\Giorgio\Documents\Python Scripts\Scacchi chessbase DB\data\chessbase\New Database_2007_2012',
           r'C:\Users\Giorgio\Documents\Python Scripts\Scacchi chessbase DB\data\chessbase\New Database_2013_2018',]
```

```python
columns_of_interest = ['Result', 'BlackElo', 'WhiteElo', 'Date', 'ECO']

for inp_dir in inp_dirs:
    df_total = pd.DataFrame(columns=columns_of_interest)
    print(inp_dir)
    file_list_json = get_file_list_json(inp_dir)
    for fjson in file_list_json:
        print('\t' + os.path.basename(fjson))
        f = open(fjson)
        dict_ = json.load(f)
        df = pd.json_normalize(dict_)
        columns_existing = list(set(columns_of_interest) & set(df.columns.to_list()))
        sub_df = df[columns_existing]# da fare join
        #sub_df = sub_df.dropna().loc[sub_df.Date != '????.??.??']
        df_total = pd.concat([df_total, sub_df], axis = 0).reset_index(drop=True)
    df_total.to_pickle(inp_dir + r'\results_elo_date.pkl')
```
